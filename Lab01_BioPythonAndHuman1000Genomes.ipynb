```python
from Bio.Seq import Seq
```

## How to do it...

### 1. "Hello World" to Biopython

```python
minha_sequencia = Seq('AGTACACTGGT')
print(minha_sequencia)
```

### 2. Complementary and complementary reverse of a sequence

```python
minha_sequencia_complement = minha_sequencia.complement()
print(minha_sequencia_complement)
```

```python
minha_sequencia_reserve_complement = minha_sequencia.reverse_complement()
print(minha_sequencia_reserve_complement)
```

### 3. Transcript

```python
dna = Seq('ATGGCCATTCGCAAGGGTGCCCGATAG')
print('DNA' + dna)
```

```python
rna = dna.transcribe()
print('RNA' + rna)
```

```python
dna2=rna.back_transcribe()
print('DNA' + dna2)
```

### 4. Translation

```python
proteina_rna = rna.translate()
print(proteina_rna)
```

```python
proteina_dna = dna.translate()
print(proteina_dna)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## How to do it...

### 1. Read the data from the file

```python
file_path = '20130502.phase3.sequence.index'
seq_data_df = pd.read_csv(file_path, delimiter='\t', header=0)
```

### 2. Basic inspection of the DataFrame

```python
display.display(seq_data_df)
```

```python
print('O dataframe tem %d colunas e %d linhas' % (seq_data_df.shape[1],seq_data_df.shape[0]) )
```

```python
print(seq_data_df.columns.tolist())
```

### Perform data cleanup

```python
# Alterar a coluna READ_COUNT para inteiro
print(seq_data_df['READ_COUNT'].dtype)
seq_data_df['READ_COUNT']=pd.to_numeric(seq_data_df['READ_COUNT'],errors='coerce').fillna(0).astype(int)
print(seq_data_df['READ_COUNT'].dtype)
```

```python
# Alterar a coluna READ_COUNT para inteiro
print(seq_data_df['BASE_COUNT'].dtype)
seq_data_df['BASE_COUNT']=pd.to_numeric(seq_data_df['BASE_COUNT'],errors='coerce').fillna(0).astype(int)
print(seq_data_df['BASE_COUNT'].dtype)
```

### Plot histogram

```python
# Contagem
center_counts = seq_data_df['CENTER_NAME'].value_counts()

# Desenho
plt.figure(figsize=(12,8))
center_counts.plot(kind='bar')
```

### Scatter Plot

```python
# Escolher só duas populações (YRI e CEU)
filtered_df = seq_data_df[(seq_data_df['POPULATION'].isin(['YRI','CEU']))]

# Desenho
plt.figure(figsize=(14,8))
scatter = sns.scatterplot(
    data = filtered_df,
    x = 'BASE_COUNT',
    y = 'READ_COUNT',
    hue = 'ANALYSIS_GROUP', # cor
    style = 'POPULATION', # forma
    palette='Set1'
    )
scatter.set_xscale('log') # escala logaritmica
scatter.set_yscale('log') # escala logaritmica
```
